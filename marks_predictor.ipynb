```python
import pandas as pd

data = {
    "Hours": [1, 2, 3, 4, 5],
    "Scores": [30, 40, 50, 60, 70]
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.linear_model import LinearRegression

X = df[["Hours"]]
y = df["Scores"]

model = LinearRegression()
model.fit(X, y)
```

```python
import pandas as pd
pred = model.predict(pd.DataFrame([[6]], columns=["Hours"]))
print("Predicted score:", pred[0])
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Hours"], df["Scores"])
plt.plot(df["Hours"], model.predict(X))
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.title("Hours vs Score Prediction")
plt.show()
```
